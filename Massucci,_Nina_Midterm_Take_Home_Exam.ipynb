```python
D = np.array([10,4.28,4,2.4,13])    # <- Put your numbers here  Diameter
C = np.array([31.42,12.57,30.16,7.54,40.82])  #                           Circumference
```

```python
plt.plot(D,C,'g.')
```

```python
x = D
y = C 
```

```python
m_guess = 0
b_guess = 0 
```

```python
x
```

```python
m_guess * x
```

```python
m_guess * x + b_guess
```

```python
y_guess = m_guess * x + b_guess
```

```python
best_loss = 1000000000
best_m = 0
best_b = 0 

for m_guess in np.linspace(0,20,100):
    for b_guess in np.linspace(0,20,100):

        print("Now Testing: ", m_guess, b_guess)
        
        y_guess = m_guess * x + b_guess

        loss = np.sum((y_guess - y)**2)

        if loss < best_loss:
            
            best_loss = loss
            best_m = m_guess
            best_b = b_guess 
```

```python
y_guess = best_m * x + best_b

plt.plot(x,y_guess,'r');
plt.plot(D,C,'g.');
plt.show()

print(best_loss,best_m,best_b)
```
